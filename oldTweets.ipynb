```python
import GetOldTweets3 as got
import numpy as np
import pandas as pd
import datetime
import time
```

```python
def get_month_tweets(month,query):
    date_date = datetime.date.today().replace(day=1).replace(month=month)
    next_date = date_date + datetime.timedelta(days=1)
    date = date_date.strftime("%Y-%m-%d")

    next_month = datetime.date.today().replace(day=28).replace(month=month)\
    + datetime.timedelta(days=4)
    last_date= (next_month - datetime.timedelta(days=next_month.day)).strftime("%Y-%m-%d")
    
    all_tweets = []
    
    while date <= last_date:
        tweetCriteria = got.manager.TweetCriteria().setQuerySearch(query)\
                .setSince(date).setUntil(next_date.strftime("%Y-%m-%d"))
        # Creation of list that contains all tweets
        tweets = got.manager.TweetManager.getTweets(tweetCriteria)
        # Creating list of chosen tweet data
        text_tweets = [[tweet.id,tweet.date,tweet.username, tweet.text,tweet.hashtags,tweet.geo,\
                        tweet.retweets,tweet.favorites,tweet.mentions,tweet.to] for tweet in tweets]
        
        all_tweets = all_tweets + text_tweets
        print(date, len(text_tweets))
        date_date = date_date + datetime.timedelta(days=1)
        date = date_date.strftime("%Y-%m-%d")
        next_date = next_date + datetime.timedelta(days=1)
    
    df = pd.DataFrame(all_tweets, columns=["id","timestamp","username","text","hashtags","geo","retweets","favorites","mentions","to"])
    df["date"] = df.timestamp.dt.date
    
    return df

def store_csv(df,month):
    df.to_csv(f'data/census_tweets_month_{month}.csv',mode = 'a', header= False, index = False)
```

```python
def get_month_tweets2(month,query,start_day=1):
    date_date = datetime.date.today().replace(day=start_day).replace(month=month)
    next_date = date_date + datetime.timedelta(days=1)

    next_month = datetime.date.today().replace(day=28).replace(month=month)\
    + datetime.timedelta(days=4)
    until_date= next_month - datetime.timedelta(days=next_month.day)
    until_date = until_date.strftime("%Y-%m-%d")
    
    today = datetime.date.today().strftime("%Y-%m-%d")
    
    while date_date.strftime("%Y-%m-%d") <= until_date and date_date.strftime("%Y-%m-%d") < today:
        incomplete = True
        while incomplete:
            try:
                tweetCriteria = got.manager.TweetCriteria().setQuerySearch(query)\
                .setSince(date_date.strftime("%Y-%m-%d")).setUntil(next_date.strftime("%Y-%m-%d"))
                # Creation of list that contains all tweets
                tweets = got.manager.TweetManager.getTweets(tweetCriteria)
                # Creating list of chosen tweet data
                text_tweets = [[tweet.id,tweet.date,tweet.username, tweet.text,tweet.hashtags,tweet.geo,\
                                tweet.retweets,tweet.favorites,tweet.mentions,tweet.to] for tweet in tweets]

                df = pd.DataFrame(text_tweets, columns=["id","timestamp","username","text","hashtags","geo","retweets","favorites","mentions","to"])
                df["date"] = df.timestamp.dt.date

                store_csv2(df,month)

                print(date_date.strftime("%Y-%m-%d"), len(text_tweets))
                date_date = date_date + datetime.timedelta(days=1)
                next_date = next_date + datetime.timedelta(days=1)
                
                incomplete = False
                
            except:
                print("Request error")
                time.sleep(60*1)
    
    df = pd.read_csv(f'data/census_tweets_month_{month}.csv')
    return df

def store_csv2(df,month):
    df.to_csv(f'data/census_tweets_month_{month}.csv',mode = 'a', header= False, index = False)
```
